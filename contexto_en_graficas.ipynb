```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

Graficamos un lienzo con una fila y dos columnas. En el primer graficamos y vs x y en el segundo x vs y:

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

Le podemos agregar un título a nuestras imágenes con set_title().

Le podemos agregar etiquetas a los ejes con set_xlabel() o set_ylabel().

Le podemos agregar una legenda a la gráfica con label dentro del plot y .legend() fuera.

Le podemos agregar notación matemática si agregamos el signo '$' antes y después del texto "$sin(X)$".

Le podemos modificar el tamaño del lienzo con figsize = (x,y) en los atributos de subplots:

```python
fig, axes = plt.subplots(1,2, figsize=(5,5))
axes[0].plot(x,y, label = "$sin(x)$")
axes[0].set_title('Relación X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()


axes[1].plot(y,x)
axes[1].set_title('Relación Y - X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

Todo esto mismo se puede usar sin la necesidad de utilizar el método orientado a objetos. Veamos:

```python
plt.plot(x, y, label = 'sin(x)')
plt.title('Este es mi Título')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc = 'lower right', bbox_to_anchor=(0.5, 0.2))
```
