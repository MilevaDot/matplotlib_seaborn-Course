```python
import matplotlib.pyplot as plt
import numpy as np
```

Es muy útil saber utilizar y generar gráficos de barras en Ciencias de Datos. Un tip adicional es que en Ciencias de Datos es usual llamarle 'variable categórica' a las variables de tipo texto.

```python
country = ['INDIA', 'JAPAN', 'MEXICO', 'COLOMBIA', 'GERMANY']
population = [1000, 800, 900, 1000, 300]
```

plt.bar genera estos gráficos de barras. Ahora, con width podemos modificar el ancho de las barras.

Con plt.xticks podemos modificar al eje x.

Dentro de xticks, rotation hará que las etiquetas roten cierto ángulo:

```python
plt.bar(country, population, width=0.7, color=['pink', 'blue', 'green', 'cyan', 'brown'])
plt.title('Gráfico de Población en Países Específicos')
plt.xlabel('Países')
plt.ylabel('Población en millones')
plt.xticks(np.arange(5), ('India', 'Japon', 'Mexico', 'Colombia', 'Alemania'), rotation=45)
plt.show()
```

Adicionalmente podemos hacer un gráfico de barras de manera horizontal:

```python
plt.barh(country, population)
plt.show()
```
