```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1, 50, 100)
```

Histrogama:

El atributo bins indica la cantidad de barras que usará para mostrar el histograma. Por defecto son 10.


```python
plt.hist(data, bins=20)
plt.show()
```

El atributo histtype indica el tipo de imagen que saldrá en el histograma:

```python
plt.hist(data, bins=20, histtype='step')
plt.show()
```

Boxplot:

Por defecto aparece de manera vertical, si hacemos vert=False se invierte.

patch_artist=True rellena de color el box.

notch=True enfoca más la visualización de los datos hacia la mediana.

```python
data = np.append(data, 200) # Con esto agregamos un dato adicional a nuestra data, dato que está muy por fuera del promedio.
plt.boxplot(data, vert=False, patch_artist=True, notch=True)
plt.show()
```

Si hacemos showfliers=False este dato lejano no se mostrará, sino solo la información aglomerada.

```python
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=False)
plt.show()
```

Scatter

Sirve mucho para observar si existe alguna tendencia o correlación entre los datos.

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N)) ** 2
colors = np.random.rand(N)
```

El atributo s nos permite darle un área o tamaño a los puntos.

El atributo c nos permite darle colores a los distintos puntos.

El atributo marker permite darle forma a estos puntos.

El atributo alpha permite visualizar mejor los puntos superpuestos, ya que da la transparencia.

```python
plt.scatter(x,y, s=area, c=colors, marker = 'p', alpha = 0.5)
plt.show()
```
