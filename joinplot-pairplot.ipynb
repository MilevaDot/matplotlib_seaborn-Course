```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

Joinplot une dos gráficas en uno para tener un gráfico más analítico.
marginal_ticks genera los ejes en los gráficos adicionales.
margianl_kws son las modificaciones en los ejes adicionales.

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist', marginal_ticks=True, marginal_kws=dict(bins=25, fill=False, multiple='dodge'))
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex', palette='dark', kind='scatter')
plt.show()
```
