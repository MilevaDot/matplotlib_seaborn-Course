```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

Un gráfico muy común para observar y graficar variables categóricas es countplot:

```python
sns.countplot(data=tips, x='day', hue='sex')
plt.show()
```

Otro gráfico muy útil es stripplot:

```python
plt.figure(figsize=(6,6))
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

Un gráfico muy similar a stripplot es swarmplot. La diferencia es que en donde más se concentren los datos, se forma una linea horizontal:

```python
plt.figure(figsize=(6,6))
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

Uno muy útil también es el boxplot:

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

Además de graficar variables categóricas, boxplot también grafica variables numéricas:

```python
sns.boxplot(data=tips, y='total_bill')
plt.show()
```

Algo interesante es que podemos combinar los gráficos:
Edit: En versiones anteriores de seaborn se usaba color='valor' en swarmplot ahora usamos palette.

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, palette='dark')
plt.show()
```

Un gráfico parecido al boxplot es el violinplot:

```python
plt.figure(figsize=(6,6))
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, palette='pastel', split=True)
plt.show()
```

violinplot también se usa en varibales numéricas:

```python
sns.violinplot(data=tips, y='total_bill')
plt.show()
```

Un gráfico muy recurrente en su uso es el catplot ya que a partir de aquí podemos ir a cualquiera de los gráficos que están arriba:

```python
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, kind='box', col='time')
plt.show()
```
