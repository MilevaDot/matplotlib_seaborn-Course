```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
```

Vamos a imprimir los estilos sobre los que podemos trabajar por default:

```python
print(plt.style.available)
```

Escojamos uno de los estilos:

```python
plt.style.use('dark_background')
```

Ahora que escogimos uno de los estilos, los lienzos y las fuentes se ver√°n distintas:

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x, x+1, 'r')
ax.plot(x, x+2, 'b')
ax.plot(x, x+3, 'y')
ax.plot(x, x+4, 'g')
ax.plot(x, x+5, 'c')
```

Ahora haremos modificaciones del tipo Matlab. Es decir, los colores y la forma de la gr√°fica entre comillas simples:

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x, x+1, 'r--')
ax.plot(x, x+2, 'b-')
ax.plot(x, x+3, 'y.-')
ax.plot(x, x+4, 'go:')
plt.show()
```

Ahora, la forma de Pyplot. Es decir, podemos acceder a estas caracter√≠sticas de manera expl√≠cita y m√°s customizable. Incluos alpha nos permite entrar a la no transparencia de la gr√°fica. Y linewidth nos permite cambiar el grosor de las gr√°ficas. linestyle permite modificar el tipo de linea que se ver√°. marker indicar qu√© deben aparecer en los puntos. markersize indica el tama√±o del punto. markerfacecolor indica el color del punto:

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x, x+1, color='green', alpha = 0.5, linewidth=5, linestyle='-', marker= 'o', markersize = 10, markerfacecolor='#CCFFCC')
ax.plot(x, x+2, color='blue', linewidth=8, linestyle='--', marker= 'x')
ax.plot(x, x+3, color='#66FF66', linewidth=3, linestyle='dashed', marker= 'v')
ax.plot(x, x+4, color='#CCFFCC', linewidth=2, linestyle=':', marker= 'P')
plt.show()
```
