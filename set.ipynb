```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x = ['A', 'B', 'C'], y = [1, 3, 2])
plt.show()
```

.set sirve para cambiar algunos par√°metros, por ejemplo, el estilo. Para el ejemplo usaremos el estilo darkgrid:

```python
sns.set(style='darkgrid')
sns.barplot(x = ['A', 'B', 'C'], y = [1, 3, 2])
plt.show()
```

O tambi√©n podemos usar el estilo dark

```python
sns.set(style='dark')
sns.barplot(x = ['A', 'B', 'C'], y = [1, 3, 2])
plt.show()
```

Se puede cambiar el atributo palette que modifica los colores de las barras:

```python
sns.set(style='dark', palette = 'muted')
sns.barplot(x = ['A', 'B', 'C'], y = [1, 3, 2])
plt.show()
```

```python
sns.set(style='dark', palette = 'deep')
sns.barplot(x = ['A', 'B', 'C'], y = [1, 3, 2])
plt.show()
```

```python
sns.set(style='dark', palette = 'dark')
sns.barplot(x = ['A', 'B', 'C'], y = [1, 3, 2])
plt.show()
```

Podemos cambiar la fuente y la escala de la fuente que usamos:

```python
sns.set(style='dark', palette = 'dark', font='Verdana', font_scale=1)
sns.barplot(x = ['A', 'B', 'C'], y = [1, 3, 2])
plt.show()
```
