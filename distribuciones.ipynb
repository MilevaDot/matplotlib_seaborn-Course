```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

histplot nos grafica una histograma:

```python
sns.histplot(data=tips, x='tip')
plt.show()
```

Podemos decidir cuántas barras se van a mostrar. Por defecto muestra 15:

```python
sns.histplot(data = tips, x = 'tip', bins = 15)
plt.show()
```

El parámetro cumulative acumula las cantidades:

```python
sns.histplot(data = tips, x = 'tip', bins = 15, cumulative=True)
plt.show()
```

El parámetro hue permite agregar otro indicativo:

```python
sns.histplot(data = tips, x = 'tip', bins = 15, cumulative=False, hue='sex')
plt.show()
```

El parámetro stat que cambia el conteo por una estadística, la que escojamos. Por defecto es el count:

```python
sns.histplot(data = tips, x = 'tip', bins = 15, cumulative=False, hue='sex', stat='frequency')
plt.show()
```

```python
sns.histplot(data = tips, x = 'tip', bins = 15, cumulative=False, hue='sex', stat='probability')
plt.show()
```

```python
sns.histplot(data = tips, x = 'tip', bins = 15, cumulative=False, hue='sex', stat='percent')
plt.show()
```

El parámetro multiple indica cómo queremos que se muestren la información gráfica. Por defecto es layer:

```python
sns.histplot(data = tips, x = 'tip', bins = 15, cumulative=False, hue='sex', stat='count', multiple='stack')
plt.show()
```

```python
sns.histplot(data = tips, x = 'tip', bins = 15, cumulative=False, hue='sex', stat='count', multiple='layer')
plt.show()
```

```python
sns.histplot(data = tips, x = 'tip', bins = 15, cumulative=False, hue='sex', stat='count', multiple='dodge')
plt.show()
```

```python
sns.histplot(data = tips, x = 'tip', bins = 15, cumulative=False, hue='sex', stat='count', multiple='fill')
plt.show()
```

Veamos un diagrama de Densidad:

```python
sns.kdeplot(data = tips, x = 'tip', hue = 'sex')
plt.show()
```

El parámetro cumulative aquí también existe:

```python
sns.kdeplot(data = tips, x = 'tip', hue = 'sex', cumulative = True)
plt.show()
```

El parámetro shade rellena de color el área bajo la curva:
Edit: Antes se usaba shade, en las nuevas versiones se escribe fill.

```python
sns.kdeplot(data = tips, x = 'tip', hue = 'sex', cumulative = False, fill = True)
plt.show()
```

bw_adjust ajusta los parámetros de la gráfica:

```python
sns.kdeplot(data = tips, x = 'tip', hue = 'sex', cumulative = False, fill = True, bw_adjust=1)
plt.show()
```

Gráfico escalonado de las proporciones de los datos x:

```python
sns.ecdfplot(data = tips, x = 'tip', hue = 'sex', stat = 'count')
plt.show()
```

Veamos nuevamente a displot:

```python
sns.displot(data = tips, x = 'tip', hue = 'sex', kind = 'kde')
plt.show()
```

Podemos hacerlo regresar a ser un histograma. displot es un graficador general ya que con este podemos ir a cualquier otro:

```python
sns.displot(data = tips, x = 'tip', hue = 'sex', kind = 'hist', multiple='stack')
plt.show()
```
